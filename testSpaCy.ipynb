```python
import spacy

nlp = spacy.load("en_core_web_lg")

frase = "She saves the king from the enemy"

doc = nlp(frase)

for token in doc:
    print(f"{token.text:<12} | POS: {token.pos_:<6} | DEP: {token.dep_}")
```

```python
nlp = spacy.load("en_core_web_lg")

frase = "She is saved by the king"

doc = nlp(frase)

for token in doc:
    print(f"{token.text:<12} | POS: {token.pos_:<6} | DEP: {token.dep_}")
```

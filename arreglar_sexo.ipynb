```python
import pandas as pd

# 1. Cargar el archivo
df = pd.read_pickle('dataset_con_genero_final.pkl')
```

```python
# --- CONFIGURA AQUÍ TUS CAMBIOS ---
fila = 1                   # La posición (0 es la primera, 1 la segunda...)
personaje = 'VOICE'        # El nombre tal cual aparece en el diccionario
nuevo_genero = 'male'      # El valor que quieres poner
# ----------------------------------
```

```python
# 2. Realizar el cambio
if personaje in df.at[fila, 'Character_Genders']:
    df.at[fila, 'Character_Genders'][personaje] = nuevo_genero
    print(f"Cambio realizado: {personaje} ahora es {nuevo_genero}")
else:
    print(f"No se encontró el personaje {personaje} en la fila {fila}")
```

```python
# 3. Guardar el resultado
df.to_pickle('dataset_con_genero_final.pkl')

# 4. Guardar los cambios en un archivo CSV
# index=False evita que se guarde una columna extra con los números de fila
df.to_csv('dataset_con_genero_final_2.csv', index=False)
```
